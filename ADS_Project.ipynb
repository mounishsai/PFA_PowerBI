```python
import pandas as pd
import numpy as np
# read datasdet using pandas
df2 = pd.read_excel('finance_final1.xlsx')
df2.head()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming you have loaded your dataset into a DataFrame called df
# Let's say the columns are 'Total Income' and 'Total Expenses'

# Extracting the independent variable (X) and dependent variable (y)
X = df2[['Total_income']]  # Features
y = df2['total_ expense'] # Target variable

# Creating a linear regression model
model = LinearRegression()

# Fitting the model
model.fit(X, y)

# Printing the coefficients
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])
```
